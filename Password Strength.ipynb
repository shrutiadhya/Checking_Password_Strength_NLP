```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('data.csv',',',error_bad_lines=False)  #if comma seperated values not properly available skip the line
data.head()
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
password_list = np.array(data)
```

```python
password_list
```

```python
import random
random.shuffle(password_list)  #shuffling randomly for robustness
```

```python
y = [rows[1] for rows in password_list]
```

```python
y
```

```python
X = [rows[0] for rows in password_list]
```

```python
X
```

```python
import seaborn as sns
```

```python
sns.set_style('whitegrid')
sns.countplot(x='strength',data=data)
```

```python
def word_divide_char(inputs):
    characters = []
    for i in inputs:
        characters.append(i)
    return characters
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(tokenizer=word_divide_char)   #instead of converting each word into token, convert each character into token
X = vectorizer.fit_transform(X)
```

```python
X
```

```python
X.shape
```

```python
vectorizer.vocabulary_
```

```python
data.iloc[0,0]
```

```python
feature_names = vectorizer.get_feature_names()
 
#get tfidf vector for first document
first_document_vector=X[0]
 
#print the scores
df = pd.DataFrame(first_document_vector.T.todense(), index=feature_names, columns=["tfidf"])  #as X is a sparse matrix convert to dense
df.sort_values(by=["tfidf"],ascending=False)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
log_model = LogisticRegression(penalty='l2',multi_class='ovr')
log_model.fit(X_train,y_train)
```

```python
log_model.score(X_test,y_test)
```

```python
## multinomial
multi_model = LogisticRegression(multi_class='multinomial',solver='newton-cg')
multi_model.fit(X_train,y_train)
multi_model.score(X_test,y_test)
```

```python
X_predict = np.array(["shruti123"])
X_predict=vectorizer.transform(X_predict)
y_pred = log_model.predict(X_predict)
print(y_pred)
```

# Xgboost classifier

```python
import xgboost as xgb
xgb_model = xgb.XGBClassifier()
```

```python
xgb_model.fit(X_train,y_train)
```

```python
xgb_model.score(X_test,y_test)
```

```python
X_predict = np.array(["shruti123@%"])
X_predict=vectorizer.transform(X_predict)
y_pred = xgb_model.predict(X_predict)
print(y_pred)
```
